### [5 points] Question 3:

Run VADER on your own tweets (see function **run_vader** from notebook **Lab2-Sentiment-analysis-using-VADER.ipynb**). You can use the code snippet below this explanation as a starting point. 
* [2.5 points] a. Perform a quantitative evaluation. Explain the different scores, and explain which scores are most relevant and why.
* [2.5 points] b. Perform an error analysis: select 10 positive, 10 negative and 10 neutral tweets that are not correctly classified and try to understand why.
````python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
vader_model = SentimentIntensityAnalyzer()
import spacy
nlp = spacy.load('en_core_web_sm')
````

````python
def vader_output_to_label(vader_output):
    """
    map vader output e.g.,
    {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.4215}
    to one of the following values:
    a) positive float -> 'positive'
    b) 0.0 -> 'neutral'
    c) negative float -> 'negative'
    
    :param dict vader_output: output dict from vader
    
    :rtype: str
    :return: 'negative' | 'neutral' | 'positive'
    """
    compound = vader_output['compound']
    
    if compound < 0:
        return 'negative'
    elif compound == 0.0:
        return 'neutral'
    elif compound > 0.0:
        return 'positive'
    
def run_vader(textual_unit, 
            lemmatize=False, 
            parts_of_speech_to_consider=None,
            verbose=0):
    """
    Run VADER on a sentence from spacy

    :param str textual unit: a textual unit, e.g., sentence, sentences (one string)
    (by looping over doc.sents)
    :param bool lemmatize: If True, provide lemmas to VADER instead of words
    :param set parts_of_speech_to_consider:
    -None or empty set: all parts of speech are provided
    -non-empty set: only these parts of speech are considered.
    :param int verbose: if set to 1, information is printed
    about input and output

    :rtype: dict
    :return: vader output dict
    """
    doc = nlp(textual_unit)
        
    input_to_vader = []

    for sent in doc.sents:
        for token in sent:

            to_add = token.text

            if lemmatize:
                to_add = token.lemma_

                if to_add == '-PRON-': 
                    to_add = token.text

            if parts_of_speech_to_consider:
                if token.pos_ in parts_of_speech_to_consider:
                    input_to_vader.append(to_add) 
            else:
                input_to_vader.append(to_add)

    scores = vader_model.polarity_scores(' '.join(input_to_vader))

    if verbose >= 1:
        print()
        print('INPUT SENTENCE', sent)
        print('INPUT TO VADER', input_to_vader)
        print('VADER OUTPUT', scores)

    return scores
    
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.0}) == 'neutral'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': 0.01}) == 'positive'
assert vader_output_to_label( {'neg': 0.0, 'neu': 0.0, 'pos': 1.0, 'compound': -0.01}) == 'negative'
````

````python
tweets = []
all_vader_output = []
gold = []

# settings (to change for different experiments)
to_lemmatize = True 
pos = set()

for id_, tweet_info in my_tweets.items():
    the_tweet = tweet_info['text_of_tweet']
    vader_output = run_vader(the_tweet, lemmatize=True) # run vader
    vader_label = vader_output_to_label(vader_output)# convert vader output to category
    
    tweets.append(the_tweet)
    all_vader_output.append(vader_label)
    gold.append(tweet_info['sentiment_label'])
    
# use scikit-learn's classification report
from sklearn.metrics import classification_report
print(classification_report(gold, all_vader_output, digits=3))
````

Quantitative evaluation:
The precision means: of all the tweets predicted as a certain class, how many were actually that class? A high precision means less false positives, in our case negative has the highest precision and positive the lowest. Recall is measured by how well the model correctly identifies tweets that truly belong to a class, high recall means few false negatives. In our case the recall is the best for the negative tweets and the worst in the neutral tweets. The f1 score is the mean of precision and recall. The support shows the true instances of each class. The most relevant scores are precision and recall, high precision and recall mean that the modal is good at identifying the sentiment.

````python
wrong = {'positive': [], 'negative': [], 'neutral': []}

for id_, tweet_info in my_tweets.items():
    the_tweet = tweet_info['text_of_tweet']
    gold = tweet_info['sentiment_label']

    vader_output = run_vader(the_tweet, lemmatize=True) # run vader
    vader_label = vader_output_to_label(vader_output) # convert vader output to category

    if vader_label != gold and len(wrong[gold]) < 10:
        wrong[gold].append((id_, the_tweet, vader_label, vader_output))

for label in ['neutral', 'negative', 'positive']:
    lst = wrong[label]
    print('\n' + '='*50)
    print(f'Misclassified (true = {label})')
    print('='*50)

    for i, (id_, the_tweet, pred, scores) in enumerate(lst, start=1):
        print(f'\n#{i} (id {id_})')
        print('TEXT:', the_tweet)
        print('GOLD:', label, 'VADER_LABEL:', pred, 'SCORES:', scores)
````

````python
# Your code here
import pathlib
from sklearn.datasets import load_files
from sklearn.metrics import classification_report
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import spacy

# Sorryyy we already finished this version before seeing we needed different code cells, so would be a lot of work to change it, hopefully this is fine
cwd = pathlib.Path.cwd()
airline_folder = cwd.joinpath('airlinetweets')
data = load_files(str(airline_folder))
vader = SentimentIntensityAnalyzer()
nlp = spacy.load('en_core_web_sm')

def vader_output_to_label(vader_output):
    c = vader_output['compound']
    if c < 0:
        return 'negative'
    elif c == 0.0:
        return 'neutral'
    else:
        return 'positive'

def run_vader(textual_unit, lemmatize=False, parts_of_speech_to_consider=None):
    doc = nlp(textual_unit)
    input_to_vader = []
    for sent in doc.sents:
        for token in sent:
            to_add = token.text
            if lemmatize:
                to_add = token.lemma_
                if to_add == '-PRON-':
                    to_add = token.text
            if parts_of_speech_to_consider:
                if token.pos_ in parts_of_speech_to_consider:
                    input_to_vader.append(to_add)
            else:
                input_to_vader.append(to_add)
    return vader.polarity_scores(' '.join(input_to_vader))

experiments = [
    ('full', False, None),
    ('full_lemma', True, None),
    ('adjectives', False, {'ADJ'}),
    ('adjectives_lemma', True, {'ADJ'}),
    ('nouns', False, {'NOUN'}),
    ('nouns_lemma', True, {'NOUN'}),
    ('verbs', False, {'VERB'}),
    ('verbs_lemma', True, {'VERB'}),
]

for name, do_lemma, pos_set in experiments:
    y_true = []
    y_pred = []
    for text, target in zip(data.data, data.target):
        text_str = text if isinstance(text, str) else text.decode('utf-8', errors='replace')
        vader_out = run_vader(text_str, lemmatize=do_lemma, parts_of_speech_to_consider=pos_set)
        pred_label = vader_output_to_label(vader_out)
        y_pred.append(pred_label)
        y_true.append(data.target_names[target])
    print('\n' + '='*60)
    print(f'Experiment: {name}  (lemmatize={do_lemma}, pos_filter={pos_set})')
    print('='*60)
    print(classification_report(y_true, y_pred, digits=3))
````

It seems like lemmatization didn't really have much help in this experiment. You can see in pretty much all experiments with the different parts of speech that each time lemmatization and no lemmatization basically had a maximum influence of around 0.05, but most of the times just a 0.01 difference in all different scores. Even sometimes having a worse score with lemmatization than without, like in the all scores in the full text experiment.
This is probably because VADER's lexicon and rule heuristics are based on full texts with punctuation, suffixes, grammar etc. so with lemmatization this full text gets reduces and removes this punctuation, suffix, etc. This explains why some scores are actually worse.

Not all parts of speech seem equally important, based on the data we can conclude that adjectives and verbs are more important than nouns. This can be explained by adjectives and nouns being important in describing thing, so they often help in showing peoples emotions with 'good', 'bad', 'happy', etc. while nouns only explain what a sentence is about, they kind of tell you the topic of the sentence, it tells you it is about a bag, or flight, etc. but it doesn't tell you exactly what is good or bad about that flight.

````python
# Your code here

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

texts = data.data

X_train_texts, X_test_texts, y_train, y_test = train_test_split(
    texts, data.target, test_size=0.2, random_state=42, stratify=data.target)

min_dfs = [2, 5, 10]
for min_df in min_dfs:
    for use_tfidf in (False, True):
        vec = CountVectorizer(min_df=min_df, stop_words='english')
        X_train_counts = vec.fit_transform(X_train_texts)
        X_test_counts = vec.transform(X_test_texts)

        if use_tfidf:
            tfidf = TfidfTransformer()
            X_train = tfidf.fit_transform(X_train_counts)
            X_test = tfidf.transform(X_test_counts)
            label = f"tfidf (min_df={min_df})"
        else:
            X_train = X_train_counts
            X_test = X_test_counts
            label = f"count (min_df={min_df})"

        clf = MultinomialNB()
        clf.fit(X_train, y_train)
        y_pred = clf.predict(X_test)

        print("\n" + "="*72)
        print(f"Experiment: {label}")
        print("="*72)
        print(classification_report(y_test, y_pred, target_names=data.target_names, digits=3))
````

The negative category performs best in every setting, in all metrics. In Lab3.3 we see that negative is the largest dataset which could be a big reason for it performing best.

The frequency threshold does seem to affect scores, but only very slightly. A higher frequency threshold gives a lower overall accuracy and f1 score. This is because raising the frequency threshold, which removes infrequent tokens, probably removes some information that could help determine if tweets are positive, negative, neutral. So words that aren't frequent, but important in determining the category are then not used in training, and could lead to some tweets being wrongly identified an thus leading to worse metric scores.

In terms of what to remove or keep: airline name tokens should be removed from the feature set — or added to a domain stop-word list — because they appear uniformly across all classes and only add noise. Numbers and standalone time tokens are similarly uninformative for sentiment and can be filtered out. Generic words like *flight*, *just*, and *get* that appear heavily in all classes should also be removed. On the other hand, strong sentiment-bearing content words such as *delay*, *cancelled*, *terrible*, and *lost* (negative) and *thank*, *great*, *amazing*, and *wonderful* (positive) should be kept, as they are the most direct indicators of a passenger's emotional evaluation and carry the most discriminative signal for the classifier.
